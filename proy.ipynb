```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("credit_risk.csv")
df.head()
```

```python
df.info()
```

```python
df = df[df.rev_util <= 1]
df.describe()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, vmin=-1, vmax=1, cmap="coolwarm")
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.2)
len(df_train), len(df_test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
y = "dlq_2yrs"
feat = list(set(df.columns).difference([y]))
feat
```

```python
for c in feat:
    sns.boxplot(x=df[c], y=df.dlq_2yrs, orient="h")
    plt.title(c)
    plt.show()
```

```python
df = df[df.late_90 < 20]
df = df[df.late_30_59 < 20]
df = df[df.late_60_89 < 20]
df = df[df.monthly_inc < 20_000]
df = df[df.debt_ratio < 1]
df["late_total"] = df.late_30_59 + df.late_60_89 + df.late_90
df["debt_load"] = df.debt_ratio * df.monthly_inc
feat.append("late_total")
feat.append("debt_load")
# df.late_30_59 = df.late_30_59 > 0
# df.late_60_89 = df.late_60_89 > 0
# df.late_90 = df.late_90 > 0
df.describe()
```

```python
for c in feat:
    sns.boxplot(x=df[c], y=df.dlq_2yrs, orient="h")
    plt.title(c)
    plt.show()
```

```python
from sklearn.decomposition import PCA, KernelPCA
```

```python
sc = StandardScaler()
pca = PCA(n_components=2)
proy = pca.fit_transform(sc.fit_transform(df[feat]))
```

```python
colors = [["r", "b"][i] for i in df[y]]
plt.scatter(proy[:, 0], proy[:, 1], color=colors, marker=".", alpha=0.1)
```

```python
sc = StandardScaler()
pca = KernelPCA(n_components=2, kernel="cosine")
sample = df
proy = pca.fit_transform(sc.fit_transform(sample[feat]))
colors = [["r", "b"][i] for i in sample[y]]
plt.scatter(proy[:, 0], proy[:, 1], color=colors, marker=".", alpha=0.1)
```

```python
df_train, df_test = train_test_split(df, test_size=0.2)
len(df_train), len(df_test)
```

```python
df_train[y].mean(), df_test[y].mean()
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(df_train[feat])
X_test = sc.transform(df_test[feat])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(max_depth=10)
tree.fit(X_train, df_train[y])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, recall_score
```

```python
y_pred = tree.predict_proba(X_train)
accuracy_score(df_train[y], y_pred[:, 1] > 0.5), recall_score(df_train[y], y_pred[:, 1] > 0.5)
```

```python
y_pred = tree.predict_proba(X_test)
accuracy_score(df_test[y], y_pred[:, 1] > 0.5), recall_score(df_test[y], y_pred[:, 1] > 0.5)
```

```python
t = np.linspace(0.1, 1, 20)
acc = [accuracy_score(df_test[y], y_pred[:, 1] > th) for th in t]
plt.plot(t, acc)
```

```python
t = np.linspace(0.1, 1, 20)
acc = [recall_score(df_test[y], y_pred[:, 1] > th) for th in t]
plt.plot(t, acc)
```

```python
roc_auc_score(df_test[y], y_pred[:, 1])
```

```python
sns.heatmap(confusion_matrix(df_test[y], y_pred[:, 1]>0.4, normalize="true"), annot=True, cmap="Blues", vmin=0, vmax=1)
plt.xlabel("Predicho")
plt.ylabel("Verdadero")
```

```python
sns.barplot(x=tree.feature_importances_, y=feat)
```
